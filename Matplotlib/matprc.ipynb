```python
import matplotlib
import matplotlib.pyplot as plt
print(matplotlib.__version__)
x =[2021,2022,2023]
y=[100,200,600]
plt.plot(x,y)
plt.xlabel("year")
plt.ylabel("count")
plt.xticks(x)
plt.show()
```

```python
#bar chart
plt.bar(x,y)
plt.xlabel("year")
plt.ylabel("count")
plt.xticks(x)
```

```python
#pie chart
plt.pie(y,labels=x,autopct="%1.1f%%",shadow=True,explode=[0,1.0,0])
```

```python
#Scatter plot
plt.scatter(x,y)
plt.xticks(x)
alpha=0.5
```

```python
#Histogram
plt.hist(x)
plt.xticks(x)
bin=10
```
